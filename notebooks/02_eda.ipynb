```python
#Imports & Data Load
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.datasets import load_breast_cancer

import matplotlib.pyplot as plt

# Plot prefs
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Load dataset as DataFrame
data = load_breast_cancer(as_frame=True)
df    = data.frame.copy()
df["target"] = data.target         # append labels

print(df.shape)
df.head()
```

```python
#Quick Overview
display(df.describe().T)

# Class balance
class_counts = df["target"].value_counts()
print("Class counts:\n", class_counts)
sns.barplot(x=class_counts.index, y=class_counts.values)
plt.title("Target Class Distribution")
plt.show()
```

```python
#Missing-Value Heatmap
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing-Value Map")
plt.show()
```

```python
#  Feature Distributions (Density Plots)
numeric_cols = df.drop(columns="target").columns
df[numeric_cols].hist(bins=30, figsize=(18,12))
plt.suptitle("Feature Histograms", fontsize=16)
plt.tight_layout()
plt.show()


```

```python
# Correlation Matrix
corr = df[numeric_cols].corr()
mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(15,12))
sns.heatmap(corr, mask=mask, cmap="coolwarm", center=0,
            linewidths=0.5, cbar_kws={"shrink": .5})
plt.title("Feature Correlation (upper-triangle hidden)")
plt.show()
```
